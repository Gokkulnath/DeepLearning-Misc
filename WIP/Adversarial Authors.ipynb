```python
def download_file(url):
    local_filename = url.split('/')[-1]
    # NOTE the stream=True parameter below
    with requests.get(url, stream=True) as r:
        r.raise_for_status()
        with open(local_filename, 'wb') as f:
            for chunk in r.iter_content(chunk_size=8192): 
                if chunk: # filter out keep-alive new chunks
                    f.write(chunk)
                    # f.flush()
    print(f"File Downloaded as {local_filename}")
    return local_filename
```

```python
import random

def search_google(string):
#     browser.execute_script("window.open('');")
    browser.get('http://www.google.com')
    search_textbox=browser.find_element_by_xpath('//input[@aria-label="Search"]')
    search_textbox.send_keys(string)
    search_textbox.send_keys(Keys.ENTER)
    links=[]
    for e in browser.find_elements_by_xpath('//div[@class="rc"]/div/a'):
#         print(e.get_attribute('href'))
        links.append(e.get_attribute('href'))
    return links
        
```

```python
uniq_auth=pd.unique(auth.index[auth>1])
# uniq_auth
```

```python
import time
from tqdm import tqdm_notebook as tqdm
browser=webdriver.Chrome()

with open('result.csv','w',encoding="utf-8") as out_fh:
    for author in tqdm(uniq_auth):
        out=search_google(author)
        time.sleep(1)
        temp_str=f"{author},"
        for link in out:
            temp_str+=f"{link},"
        temp_str.encode("utf-8")
        out_fh.writelines(temp_str+'\n')
        
    
```
